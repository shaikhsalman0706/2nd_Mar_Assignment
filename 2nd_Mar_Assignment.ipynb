### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#### Matplotlib is a popular open-source data visualization library for Python. It provides a wide range of tools for creating high-quality, customizable visualizations in a variety of formats. Matplotlib is used extensively in the scientific community and is also used by data analysts and machine learning practitioners for data exploration, analysis, and communication.

#### Matplotlib's Pyplot module is a sublibrary that provides a simple interface for creating a wide variety of plots. Some of the commonly used plots that can be created using the Pyplot module of Matplotlib are:

###### Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. This type of plot is useful for showing trends over time.

###### Scatter Plot: A scatter plot is a graph that displays data as a collection of points. This type of plot is useful for showing the relationship between two variables.

###### Bar Plot: A bar plot is a graph that displays data as a series of bars. This type of plot is useful for comparing the values of different categories.

###### Histogram: A histogram is a graph that displays the distribution of a dataset. This type of plot is useful for showing the frequency of values in a dataset.

###### Pie Chart: A pie chart is a graph that displays data as a set of slices of a circular pie. This type of plot is useful for showing the proportion of different categories in a dataset.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
###### import numpy as np
###### np.random.seed(3)
###### x = 3 + np.random.normal(0, 2, 50)
###### y = 3 + np.random.normal(0, 2, len(x))
###### Note: Also add title, xlabel, and ylabel to the plot.

#### A scatter plot is a type of plot that displays the relationship between two variables. It is created by plotting one variable on the x-axis and another variable on the y-axis, and then using dots or other markers to represent the data points.

#### Here is an example code that generates data for x and y, and then plots a scatter plot using Matplotlib's Pyplot module:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title and axis labels
plt.title("Scatter Plot of X and Y")
plt.xlabel("X")
plt.ylabel("Y")

# Show the plot
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
### Use the following data:
###### import numpy as np
###### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
###### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
###### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
###### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows us to create a grid of subplots, where each subplot can display a different visualization. This is useful for comparing multiple visualizations side by side or for displaying related data.

#### Here is an example code that uses the subplot() function to create four line plots:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data for the four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot each line plot in a different subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a main title to the figure
fig.suptitle('Four Line Plots')

# Display the figure
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
### import numpy as np
### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
### profit = np.array([3000, 8000, 1000, 10000])

#### A bar plot is a type of plot that represents data using rectangular bars. The height or length of each bar corresponds to the value of the data being plotted, and the bars can be arranged vertically or horizontally depending on the orientation of the plot. Bar plots are commonly used to compare different categories of data and to show trends over time.

#### Here is an example code that uses Matplotlib's bar() function to create a vertical bar plot and a horizontal bar plot using the given data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
### box1 = np.random.normal(100, 10, 200)
### box2 = np.random.normal(90, 20, 200)

#### A box plot, also known as a box-and-whisker plot, is a type of plot used to visualize the distribution of a dataset. It displays the median, quartiles, and outliers of the data, making it useful for identifying skewness, outliers, and the spread of the data.

#### In a box plot, the box represents the interquartile range (IQR), which is the range between the 25th and 75th percentiles of the data. The line inside the box represents the median, or the 50th percentile. The "whiskers" extend from the box to the minimum and maximum values that are not considered outliers. Outliers are shown as individual points beyond the whiskers.

#### Here is an example code that uses Matplotlib's boxplot() function to create a box plot using the given data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Create the box plot
plt.boxplot(data)
plt.title('Box plot')
plt.xlabel('Box')
plt.ylabel('Value')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
